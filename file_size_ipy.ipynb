```python
import os
from pprint import pprint
from matplotlib import pyplot as plt


""" Возвращает размер указанного файла или каталога """

def get_size(start_path = '../'):
    total_size = 0

    if os.path.isfile(start_path):
        return os.path.getsize(start_path)

    elif os.path.isdir(start_path):
        for dirpath, dirnames, filenames in os.walk(start_path):
            for f in filenames:
                fp = os.path.join(dirpath, f)
                # skip if it is symbolic link
                if not os.path.islink(fp):
                    # print(fp, os.path.getsize(fp))
                    total_size += os.path.getsize(fp)

    return total_size


def get_size_in_dir(abs_path ='C:/Users/corsair/'):
    list_items = os.listdir(abs_path)
    dict_items = {}

    for elem in list_items:
        size_ = get_size(start_path = abs_path + elem)
        dict_items[elem] = size_

    return dict_items
```

```python
list_items = os.listdir('../')
list_sizes = []
dict_items = {}

# pprint(list_dirs)
for elem in list_items:
    size_ = get_size(start_path = '../' + elem)
    size_ /= 1024
    
    list_sizes.append(size_)
    dict_items[elem] = size_
    
    print(elem.ljust(18), str(size_)[:6], 'kb')
    
plt.figure(figsize=(12, 12))
plt.barh(list_items, list_sizes)
plt.show()
```
